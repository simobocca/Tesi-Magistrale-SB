### Aggregazione per nshots

```python
def team_avg_users_nshot(team):
    users = db.users_in_teams.find({'team':team})
    nshots = []
    for u in users:
        nshots.append(uf.user_nshots(u['username']))
        
        
    if users.count() != 0:
        return np.average(nshots)/float(users.count())
    else:
        return np.average(nshots)


def team_max_users_nshot(team):
    users = db.users_in_teams.find({'team':team})
    nshots = []
    for u in users:
        nshots.append(uf.user_nshots(u['username']))
        
    if users.count() != 0:
        return np.max(nshots)/float(users.count())
    else:
        return np.max(nshots)

def team_var_users_nshot(team):
    users = db.users_in_teams.find({'team':team})
    nshots = []
    for u in users:
        nshots.append(uf.user_nshots(u['username']))
        
    if users.count() != 0:
        return np.max(nshots)/float(users.count())
    else:
        return np.max(nshots)

def team_med_users_nshot(team):
    users = db.users_in_teams.find({'team':team})
    nshots = []
    for u in users:
        nshots.append(uf.user_nshots(u['username']))
        
    if users.count() != 0:
        return np.median(nshots)/float(users.count())
    else:
        return np.median(nshots)
```
