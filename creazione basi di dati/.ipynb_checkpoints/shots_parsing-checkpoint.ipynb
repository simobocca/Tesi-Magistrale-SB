```python
import time
from datetime import datetime


json_shot[0]['username'] = desc_user[771252]


json_shot[0].pop('attachments_count', None)
json_shot[0].pop('attachments_url', None)
json_shot[0].pop('buckets_count', None)
json_shot[0].pop('buckets_url', None)
json_shot[0].pop('comments_url', None)
json_shot[0].pop('html_url', None)
json_shot[0].pop('images', None)
json_shot[0].pop('normal', None)
json_shot[0].pop('teaser', None)
json_shot[0].pop('likes_url', None)
json_shot[0].pop('projects_url', None)
json_shot[0].pop('rebounds_count', None)
json_shot[0].pop('rebounds_url', None)

d = datetime.strptime(json_shot[0]['created_at'], '%Y-%m-%dT%H:%M:%SZ')
unixtime = time.mktime(d.timetuple())
json_shot[0]['created_at'] = int(unixtime)

d = datetime.strptime(json_shot[0]['updated_at'], '%Y-%m-%dT%H:%M:%SZ')
unixtime = time.mktime(d.timetuple())
json_shot[0]['updated_at'] = int(unixtime)
```
