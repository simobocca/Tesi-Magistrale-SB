```python
import numpy as np
import pandas as pd
import json
import pygraphviz as pgv
import collections
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G_users = pgv.AGraph()
G_users.read('/home/simo/UnInfo/Tesi Magistrale/test/userFollowersGraph.dot')

G_teams = pgv.AGraph()
G_teams.read('/home/simo/UnInfo/Tesi Magistrale/test/teamsGraph.dot')
```

```python
print 'user nodi: ', len(G_users.nodes())
print 'user archi: ', len(G_users.edges())
print '\n'
print 'teams nodi: ', len(G_teams.nodes())
print 'teams archi: ', len(G_teams.edges())
```

```python
users_nxG = nx.Graph(G_users)
teams_nxG = nx.Graph(G_teams)
users_centralities = nx.degree_centrality(users_nxG)
teams_centralities = nx.degree_centrality(teams_nxG)
```

```python
# user con centralità massima 
max = 0
elem = ''
for u in users_centralities:
    if users_centralities[u] > max:
        max = users_centralities[u]
        elem = u
print elem, max

# team con centralità massim
max = 0
elem = ''
for u in teams_centralities:
    if teams_centralities[u] > max:
        max = teams_centralities[u]
        elem = u
print elem, max
```

```python
print 'centralità media per users', np.average(users_centralities.values()) 
print 'centralità media per teams', np.average(teams_centralities.values()) 
```

```python
def degree_distrib(G, r, title):
    degree_sequence = sorted([d for d in G.degree()], reverse=True)  # degree sequence
    #print "Degree sequence", degree_sequence
    
    print 'grado massimo per '+title, np.max(degree_sequence)
    print 'media dei gradi per '+title, np.average(degree_sequence)
    
    (values,counts) = np.unique(degree_sequence, return_counts=True)
    ind = np.argmax(counts)
    print 'moda dei gradi per '+title, values[ind]

    degreeCount = collections.Counter(degree_sequence)
    deg, cnt = zip(*degreeCount.items())

    fig, ax = plt.subplots()
    fig.set_size_inches(20,10)

    plt.bar(deg, cnt, align='edge', width=0.80, color='b')

    plt.title("Degree Histogram "+title)
    plt.ylabel("Count")
    plt.xlabel("Degree")
    
    if not r is None:
        ax.set_xticks([d*10 for d in r])
        ax.set_xticklabels([d*10 for d in r])
    else:
        ax.set_xticklabels([0,1,10,100,1000,100000])
    
    plt.savefig(title+'.png')

    plt.show()
```

```python
degree_distrib(G_users, None, 'Users')
degree_distrib(G_teams, range(0,5), 'Teams')
```
